```python
import os
import shutil
import json

# Base project directory
base_dir = "C:/Users/ningw/Desktop/asp/EnergyStorageOptimisationASP"

# Source and destination directories
src = os.path.join(base_dir, "Data")
dst = os.path.join(base_dir, "Data_Extreme_Solar")

# Copy Data to Data_Extreme_Solar (overwrite if exists)
shutil.copytree(src, dst, dirs_exist_ok=True)

# Loop through all JSON files in the new directory
for filename in os.listdir(dst):
    if filename.endswith(".json"):
        file_path = os.path.join(dst, filename)

        # Load JSON data
        with open(file_path, "r") as f:
            data = json.load(f)

        # Update schedule_input entries
        if "schedule_input" in data:
            for entry in data["schedule_input"]:
                entry["production_forecast_ac"] = 1500
                entry["production_forecast_dc"] = 1500

        # Save back to the same file
        with open(file_path, "w") as f:
            json.dump(data, f, indent=4)

        print(f"Updated {filename}")
```
